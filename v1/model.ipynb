## define model


```python
model = BloodType(10000, timesteptype='m')
```

```python
# iterations
for i in tqdm(range(100)):
    model.step()
model.printState()

for i in tqdm(range(100)):
    model.step(bt_mutation='A', mutations=10)
    model.step(bt_mutation='B', mutations=10)

model.printState()

for i in tqdm(range(100)):
    model.step()
model.printState()

model.setFitness(fitness = {
        'O': 100,
        'A': 1,
        'B': 5,
        'AB': 20
    })
model.setDeathRate(value=0.8)

for i in tqdm(range(36)):
    model.step()
model.printState()

model.setFitness(fitness = {
        'O': 1,
        'A': 1,
        'B': 1,
        'AB': 1
    })
model.setDeathRate(value=0.01)


for i in tqdm(range(5000)):
    model.step()
model.printState()
```

```python
# model.plotSize()
model.plotSize()
model.plotSize(cumulativ=True)
# model.plotSex(ratio=True)
model.plotBtPt()
```
